Exercício 9 - Cálculo de Tokens em Texto Longo com API Gemini

Implemente um notebook que use a API Gemini para calcular a quantidade de tokens necessários para processar um texto de 5.000 palavras. Baseie-se no modelo de tokenização utilizado por Gemini e explique como a quantidade de tokens influencia o custo e o desempenho da interação com LLMs em textos longos.

```python
# Para modelos do Gemini, um token equivale a cerca de quatro caracteres. 100 tokens equivalem a cerca de 60-80 palavras em inglês.

import google.generativeai as genai

genai.configure(api_key="AIzaSyDBEsTsQs1xHPaYQ3DJZn8cXUd8RiMOtjg")
modelo_info = genai.get_model('models/gemini-1.5-flash')

limite_token_input = modelo_info.input_token_limit

# Se 100 tokens equivalem a cerca de 60-80 palavras em inglês. utilizando 70 palavras a cada 100 tokens obtemos:
utilizacao_tokens = round(((5000) / (70/100)), 2)

print(f"O limite de tokens de input na API do Gemini é de: {limite_token_input}, considerando o enunciado se for 5000 palavras, obtemos uma utilização de toknes na quantidade de: {utilizacao_tokens}, ou seja, restando {limite_token_input - utilizacao_tokens} tokens")

# Quanto maior o prompt utilizado para o modelo maior a utilização de tokens, com isso o desempenho diminui e aumenta o custo, o custo de uma chamada para a API Gemini é determinado em parte pelo número de tokens de entrada e saída. Por isso, saber como contar os tokens pode ser útil.
```

Exercício 12 - Sistema de Envio de Emails com Dados Turísticos em JSON

Queremos montar um sistema que envie emails com os pontos turísticos de cidades num roteiro de viagem. Com base no Exercício 11, altere o prompt para estruturar a saída em formato JSON, com descrições e número de visitantes. Crie um notebook que use a biblioteca pandas para converter esse JSON para dataframe e plot um único gráfico de barras horizontais com todos os pontos turísticos, destacando as cores das barras de acordo com a cidade do ponto turístico (Coliseu, em Roma deve ter uma cor diferente que Parthenon, em Atenas). Coloque as legendas no gráfico.

```python
import yaml
import google.generativeai as genai
import pandas as pd
import json
import matplotlib.pyplot as plt

with open("ex_12.yaml", "r") as file:
    dados = yaml.safe_load(file)

genai.configure(api_key="AIzaSyDBEsTsQs1xHPaYQ3DJZn8cXUd8RiMOtjg")
modelo = genai.GenerativeModel('gemini-1.5-flash')

def resposta_lista_prompts(dados):
    lista_respostas = []
    lista_prompts = list(dados['roteiro'].values())
    for prompts in lista_prompts:
        resposta = modelo.generate_content(prompts)
        lista_respostas.append(resposta.text)
    return lista_respostas

lista_resposta_modelo = resposta_lista_prompts(dados)

atena = lista_resposta_modelo[0]
roma = lista_resposta_modelo[1]

json_atena = json.loads(atena)
json_roma = json.loads(roma)

dados_atena = json_atena["Atenas"]
dados_roma = json_roma['Roma']


df_atena = pd.DataFrame(dados_atena)
df_atena['Cidade'] = "Atenas"

df_roma = pd.DataFrame(dados_roma)
df_roma['Cidade'] = "Roma"

df = pd.concat([df_roma, df_atena], axis=0)


#Plot do gráfico
plt.figure(figsize=(10, 6))
posicoes = range(len(df))
cores = ['blue' if cidade == 'Atenas' else 'red' for cidade in df['Cidade']]

plt.bar(posicoes, [1]*len(df), color=cores)
plt.xticks(posicoes, df['ponto_turistico'], rotation=45)
plt.title('Pontos Turísticos por Cidade')
plt.ylabel('Contagem')
plt.tight_layout()
plt.show()
```
