Exerc√≠cio 12 - Sistema de Envio de Emails com Dados Tur√≠sticos em JSON

Queremos montar um sistema que envie emails com os pontos tur√≠sticos de cidades num roteiro de viagem. Com base no Exerc√≠cio 11, altere o prompt para estruturar a sa√≠da em formato JSON, com descri√ß√µes e n√∫mero de visitantes. Crie um notebook que use a biblioteca pandas para converter esse JSON para dataframe e plot um √∫nico gr√°fico de barras horizontais com todos os pontos tur√≠sticos, destacando as cores das barras de acordo com a cidade do ponto tur√≠stico (Coliseu, em Roma deve ter uma cor diferente que Parthenon, em Atenas). Coloque as legendas no gr√°fico.

```python
import yaml
import google.generativeai as genai
import pandas as pd
import json
import matplotlib.pyplot as plt

with open("ex_12.yaml", "r") as file:
    dados = yaml.safe_load(file)

genai.configure(api_key="AIzaSyDBEsTsQs1xHPaYQ3DJZn8cXUd8RiMOtjg")
modelo = genai.GenerativeModel('gemini-1.5-flash')

def resposta_lista_prompts(dados):
    lista_respostas = []
    lista_prompts = list(dados['roteiro'].values())
    for prompts in lista_prompts:
        resposta = modelo.generate_content(prompts)
        lista_respostas.append(resposta.text)
    return lista_respostas

lista_resposta_modelo = resposta_lista_prompts(dados)

atena = lista_resposta_modelo[0]
roma = lista_resposta_modelo[1]

json_atena = json.loads(atena)
json_roma = json.loads(roma)

dados_atena = json_atena["Atenas"]
dados_roma = json_roma['Roma']


df_atena = pd.DataFrame(dados_atena)
df_atena['Cidade'] = "Atenas"

df_roma = pd.DataFrame(dados_roma)
df_roma['Cidade'] = "Roma"

df = pd.concat([df_roma, df_atena], axis=0)


#Plot do gr√°fico
plt.figure(figsize=(10, 6))
posicoes = range(len(df))
cores = ['blue' if cidade == 'Atenas' else 'red' for cidade in df['Cidade']]

plt.bar(posicoes, [1]*len(df), color=cores)
plt.xticks(posicoes, df['ponto_turistico'], rotation=45)
plt.title('Pontos Tur√≠sticos por Cidade')
plt.ylabel('Contagem')
plt.tight_layout()
plt.show()
```
